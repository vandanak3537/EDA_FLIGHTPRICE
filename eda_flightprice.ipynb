```python
##importing all the libraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
##getting the basic infor of te data
df.info()
```

```python
df.describe() ##only about price cuz its int
```

```python
##Feature Engineering
#we are adding date,month and year as new feature from date of journey feature
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df.head()
```

```python
#convering date into int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
##Drop date of journey column cuz its not required
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
##splitting the month in arrival time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
#Splitting the arrival_time into hour and minute features
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True, errors='ignore')
```

```python
df.head(2)
```

```python
##Now splitting the departure time
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time', axis=1, inplace=True, errors='ignore')
```

```python
df.head()
```

```python
##using categorical features
df['Total_Stops'].unique()
```

```python
##we are replacing the values with something meaningful (0,1,2,3,..)
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4,np.nan:1})
```
